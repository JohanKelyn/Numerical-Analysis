# Fixed-Point Iteration

The goal of the fixed-point iteration is to find a "x-value" such that our expresion f(x) = 0 can be expressed as g(x) = x.

<strong>Definition: </strong> The real number r is a fixed-point of the function g if g(r) = r

```python
# Fixed-Point Algorithm

def fpi(x0,g,tol):
    """
    Input: x0 is the initial guess, g is the funtion, and tol refers to the number of steps for the algorithm
    Output: list of x values
    
    """
    x = list()
    x_old = x0
    x.append(x0)
    error = 1 
    while error > tol:
        x_new = g(x_old) # finding the new value of x
        error = np.abs(x_new - x_old) # calculating the error
        x.append(x_new) # adding our new x value to the list
        x_old = x_new # assigning the new value to the old value 
        if error > 100:
            break
    return x
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

<strong>Example 1: </strong> Find the root of the following function f(x) = x^3-2x-2 using the Fixed-Point Iteration with a presition of 8 digits. 

In order to solve this problem, we will solve for x to express our equation f(x) = 0 into x = g(x).
* x = (2x+2)^(1/3)

```python
g1 = lambda x: (2*x+2)**(1/3)
x0 = 0.5
tol = 0.00000001
sol1 = fpi(x0,g1,tol)
print(sol1)
```

Let's graph our function to see if our solution is a good approximation:

```python
x = np.linspace(0,2)
y = x**3 - 2*x - 2

plt.plot(x, y, 'mo-')
plt.axhline(0,linewidth=1, color='k')
plt.title("Example 1")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.show()
```

<strong>Example 2: </strong> Find the root of the following function f(x) = e^x + x - 7 using the Fixed-Point Iteration with a presition of 8 digits. 

In order to solve this problem, we will solve for x to express our equation f(x) = 0 into x = g(x).

```python
g2 = lambda x: np.log(7-x)
x0 = 0.5
tol = 0.00000001
sol2 = fpi(x0,g2,tol)
print(sol2)

# Ploting the graph to verify our solution
y2 = np.exp(x) + x - 7
plt.plot(x, y2, 'r--')
plt.title("Example 2")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.axhline(0,linewidth=1, color='g')
plt.show()
```
