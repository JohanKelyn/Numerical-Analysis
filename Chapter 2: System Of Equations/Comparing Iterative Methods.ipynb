# Comparing Iterative Methods

<strong>Example 1:</strong>Use the Jacobi, Gauss-Seidel, and the SOR methods to solve the following system of equations and compare their number of steps to converge to a solution.

Let's define $A_{i,i} = 3$, $A_{i,i+1} = -1$ and $A_{i+1,i} = -1$ for $n = 5$, and $b = [2, 1, 1, ..., 1, 2]$

```python
%run jacobi.ipynb
%run gauss_seidel.ipynb
%run sor.ipynb
%matplotlib inline
```

```python
n = 5
A = np.identity(n)*3

for i in range(n-1):
    A[i,i+1] = -1
    A[i+1,i] = -1

b = np.ones((n,1))
b[0] = 2
b[-1] = 2
print(A)
print(b)
```

```python
tol = 10**-6
w = 1.13
[x1, step1, err1] = jacobi(A,b,tol)
[x2, step2, err2] = gauss_seidel(A,b,tol)
[x3, step3, err3] = sor(A,b,tol,w)
```

```python
plt.figure(figsize=(15,5))
plt.plot(err1,'b')
plt.plot(err2,'r')
plt.plot(err3,'g')
plt.title("Comparing Iterative Methods")
plt.xticks(np.arange(0,len(err1)))
plt.xlabel("step")
plt.ylabel("error")
plt.legend(['Jacobi','Gauss','SOR'])
```

```python
print("Number of steps in Jacobi: ")
print(step1)
print("Number of steps in Gauss-Seidel: ")
print(step2)
print("Number of steps in SOR: ")
print(step3)
```
